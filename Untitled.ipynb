## Standard Scalar

Standardizing means subtracting off each feature column's mean and then dividing by its standard deviation so that the resulting variable has mean 0 and standard deviation 1. This ensures that the variables are penalized fairly with respect to one another.

```python
airbnb_df = airbnb_df.drop('price', axis=1)
```

```python
x = airbnb_df.drop('price_log', axis=1)
y = airbnb_df['price_log']
```

```python
scaler = StandardScaler()
airbnb_x = scaler.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(airbnb_x, y, test_size=0.3,random_state=42)
```

```python
#Linear Regression Model
lr = LinearRegression()
lr.fit(X_train, y_train)

pred_train_lr= lr.predict(X_train)
pred_test_lr= lr.predict(X_test)

print('RMSE Train: %f'% np.sqrt(mean_squared_error(y_train,pred_train_lr)))
print('RMSE Test: %f'% np.sqrt(mean_squared_error(y_test,pred_test_lr))) 
print('R2 Train: %f' % r2_score(y_train, pred_train_lr))
print('R2 Test: %f' % r2_score(y_test, pred_test_lr))
```

```python
#Lasso Model
lasso = Lasso(alpha=10)
lasso.fit(X_train, y_train) 
pred_train_lasso= lasso.predict(X_train)
pred_test_lasso= lasso.predict(X_test)

print('RMSE Train: %f'%np.sqrt(mean_squared_error(y_train,pred_train_lasso)))
print('RMSE Test: %f'% np.sqrt(mean_squared_error(y_test,pred_test_lasso))) 
print('R2 Train: %f' % r2_score(y_train, pred_train_lasso))
print('R2 Test: %f' % r2_score(y_test, pred_test_lasso))
```

```python
#Ridge Model
ridge = Ridge(alpha=10)
ridge.fit(X_train, y_train) 
pred_train_ridge= ridge.predict(X_train)
pred_test_ridge= ridge.predict(X_test)

print('RMSE Train: %f'% np.sqrt(mean_squared_error(y_train,pred_train_ridge)))
print('RMSE Test: %f'% np.sqrt(mean_squared_error(y_test,pred_test_ridge))) 
print('R2 Train: %f' % r2_score(y_train, pred_train_ridge))
print('R2 Test: %f' % r2_score(y_test, pred_test_ridge))
```

```python
#Elastic Net
model_en = ElasticNet(alpha = 10)
model_en.fit(X_train, y_train) 
pred_train_en= model_en.predict(X_train)
pred_test_en= model_en.predict(X_test)

print('RMSE Train: %f'% np.sqrt(mean_squared_error(y_train,pred_train_en)))
print('RMSE Test: %f'% np.sqrt(mean_squared_error(y_test,pred_test_en)))
print('R2 Train: %f' % r2_score(y_train, pred_train_en))
print('R2 Test: %f' % r2_score(y_test, pred_test_en))
```

Ideally, would prefer a low RMSE and a high R-squared for a model. The RMSE is relatively high across all these models which measures the average magnitude of errors while R-squared represents the coefficient of determination. Here the Linear Regression and Ridge Models have similar RMSE and R-squares.
